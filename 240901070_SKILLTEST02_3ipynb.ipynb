```python
3)Write a Python script for Flight Data Analysis:
Problem Statement:
You have a CSV file named 'flight_data.csv' containing flight information such as origin,
destination, and departure time.
Perform the following tasks:

a. Display the summary statistics for departure delays,
b. Calculate the average delay for each airline,
c. Identify the most common departure and arrival destinations,
d. Visualize the distribution of departure delays using a box plot.
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Flight ID": ["F101","F102","F103","F104","F105","F106","F107","F108","F109","F110"],
    "Airline": ["IndiGo","Air India","SpiceJet","IndiGo","Vistara","Air India","SpiceJet","IndiGo","Vistara","Air India"],
    "Origin": ["Chennai","Delhi","Chennai","Mumbai","Delhi","Bangalore","Mumbai","Chennai","Delhi","Mumbai"],
    "Destination": ["Mumbai","Bangalore","Delhi","Chennai","Mumbai","Chennai","Delhi","Bangalore","Chennai","Bangalore"],
    "Departure Delay": [10,25,5,0,15,30,20,0,12,18]
}
df = pd.DataFrame(data)
# Save CSV
df.to_csv('flight_data.csv', index=False)
print("CSV file created successfully!\n")

df = pd.read_csv('flight_data.csv')
print("Flight Data:\n", df, "\n")

print("Summary Statistics (Departure Delay):\n")
print(df['Departure Delay'].describe(), "\n")

avg_delay = df.groupby('Airline')['Departure Delay'].mean()
print("Average Delay per Airline:\n")
print(avg_delay, "\n")

most_common_origin = df['Origin'].value_counts().idxmax()
most_common_destination = df['Destination'].value_counts().idxmax()
print("Most Common Origin:", most_common_origin)
print("Most Common Destination:", most_common_destination, "\n")

plt.figure()
df['Departure Delay'].plot(kind='box')

plt.title('Distribution of Departure Delays')
plt.ylabel('Delay (minutes)')

plt.tight_layout()
plt.show()
```
