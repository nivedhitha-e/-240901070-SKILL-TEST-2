```python
6) Write a program to simulate and analyze daily stock prices over one year and generate
the stock prices, calculate some basic metrics, and visualize the data with additional
features like moving averages and visualize the same.
```

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate date range (1 year)
dates = pd.date_range(start='2025-01-01', periods=365)

# Simulate stock prices using random walk
np.random.seed(0)
price = 100
prices = []

for i in range(365):
    change = np.random.normal(0, 1)
    price += change
    prices.append(price)

# Create DataFrame
df = pd.DataFrame({'Date': dates, 'Price': prices})
df.set_index('Date', inplace=True)

# Basic metrics
print(df.describe())

# Moving averages
df['MA_7'] = df['Price'].rolling(7).mean()
df['MA_30'] = df['Price'].rolling(30).mean()

# Plot
plt.figure()
plt.plot(df['Price'], label='Price')
plt.plot(df['MA_7'], label='7-day MA')
plt.plot(df['MA_30'], label='30-day MA')

plt.title('Stock Price Simulation')
plt.legend()
plt.show()
```
