```python
Write a program to calculate Euclidean distance for two points in 2D space and Dot product
for two matrix using the mathematical formulas.
```

```python
import math
# 1. Euclidean Distance (2D)
# Formula: √((x2 - x1)^2 + (y2 - y1)^2)

def euclidean_distance(x1, y1, x2, y2):
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distance
# Example points
x1, y1 = 2, 3
x2, y2 = 6, 7
dist = euclidean_distance(x1, y1, x2, y2)
print("Euclidean Distance:", dist)
# 2. Dot Product of Two Matrices
def matrix_dot_product(A, B):
    # Number of rows in A and columns in B
    result = []
    # Check if multiplication is possible
    if len(A[0]) != len(B):
        print("Matrix multiplication not possible")
        return None

    # Perform multiplication
    for i in range(len(A)):
        row = []
        for j in range(len(B[0])):
            sum_val = 0
            for k in range(len(B)):
                sum_val += A[i][k] * B[k][j]
            row.append(sum_val)
        result.append(row)

    return result
# Example matrices
A = [
    [1, 2],
    [3, 4]
]
B = [
    [5, 6],
    [7, 8]
]
dot_product = matrix_dot_product(A, B)
print("\nMatrix Dot Product:")
for row in dot_product:
    print(row)
```
