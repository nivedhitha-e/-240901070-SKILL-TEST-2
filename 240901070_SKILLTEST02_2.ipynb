```python
2) You are given a CSV file named 'student_scores.csv' containing scores of students in
various subjects.
Perform the following tasks:
a. Display the summary statistics for each subject,
b. Calculate the average score for each student,
c. Identify students who scored below 60 in more than two subjects.
```

```python
import pandas as pd
df = pd.read_csv('student_scores.csv')

# Display first few rows
print("Student Data:\n", df.head())
print("\nSummary Statistics:\n")
print(df[['Maths', 'Physics', 'Chemistry']].describe())

df['Average'] = df[['Maths', 'Physics', 'Chemistry']].mean(axis=1)
print("\nAverage Score of Each Student:\n")
print(df[['Student\r\nID', 'Name', 'Average']])

# Count subjects where score < 60
df['Low_Score_Count'] = (df[['Maths', 'Physics', 'Chemistry']] < 60).sum(axis=1)

# Filter students
low_performers = df[df['Low_Score_Count'] > 2]

print("\nStudents scoring below 60 in more than two subjects:\n")
print(low_performers[['Student\r\nID', 'Name']])
```
